b)	Write a function to calculate the standard deviation of the values in the list create. (3pt)

$s$: standard deviation, $x$: element in the list, $\bar{x}$: average of the elements, n: number of elements. <br> **DO NOT use any pre-defined functions to find the standard deviation (e.g., sum, count, etc)**.

```python

def get_std_dev(ls):
    n = len(ls)
    mean = sum(ls) / n
    var = sum((x - mean)**2 for x in ls) / n
    std_dev = var ** 0.5
    return std_dev

ls = [1,2,3]
get_std_dev(ls)
```
